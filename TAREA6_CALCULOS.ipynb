```python
import numpy as np
import matplotlib.pyplot as plt

# Estilo pastel de seaborn
plt.style.use('seaborn-v0_8-pastel')

# Rango de x
x = np.linspace(-0.3, 0.3, 1000)

# Función original y aproximaciones de Taylor
y_original = 1 / (25 * x**2 + 1)
y_taylor_0 = np.ones_like(x)
y_taylor_2 = 1 - 25 * x**2
y_taylor_4 = 1 - 25 * x**2 + 312.5 * x**4
y_taylor_6 = 1 - 25 * x**2 + 312.5 * x**4 - ((4312500 / 720) * x**6)

# Crear figura
plt.figure(figsize=(12, 7))

# Graficar funciones con colores suaves y agradables
plt.plot(x, y_original, label='f(x) = 1 / (25x² + 1)', color='teal', linewidth=2)
plt.plot(x, y_taylor_0, label='Taylor orden 0', color='lightgray', linestyle='--', linewidth=2)
plt.plot(x, y_taylor_2, label='Taylor orden 2', color='skyblue', linestyle='-.', linewidth=2)
plt.plot(x, y_taylor_4, label='Taylor orden 4', color='salmon', linestyle=':', linewidth=2)
plt.plot(x, y_taylor_6, label='Taylor orden 6', color='orchid', linestyle='--', linewidth=2)

# Personalización
plt.ylim([0, 1.1])
plt.title('Aproximaciones de la Serie de Taylor', fontsize=16)
plt.xlabel('x', fontsize=13)
plt.ylabel('y', fontsize=13)
plt.axhline(0, color='gray', linewidth=0.8, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.8, linestyle='--')
plt.legend(fontsize=10)
plt.grid(True, linestyle=':', alpha=0.5)
plt.tight_layout()


# Mostrar
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Rango de x
x = np.linspace(-0.6, 0.6, 1000)
y_original = 1 / (25 * x**2 + 1)
y_lagrange_o3 = (
    4/29 * ((x*(x-0.5))/0.5) -
    ((x+0.5)*(x-0.5))/0.25 +
    4/29 * (x*(x+0.5))/0.5
)
y_lagrange_o5 = (
    4/29 * (((x+0.2)*x*(x-0.2)*(x-0.5))/0.105) -
    0.5 * (((x+0.5)*x*(x-0.2)*(x-0.5))/0.0168) +
    (((x+0.5)*(x+0.2)*(x-0.2)*(x-0.5))/0.01) -
    0.5*(((x+0.5)*(x+0.2)*x*(x-0.5))/0.0168) +
    4/29 * (((x+0.5)*(x+0.2)*x*(x-0.2))/0.105)
)

x_points = np.array([-0.5,-0.2,0,0.2,0.5])
y_points = np.array([4/29,0.5,1,0.5,4/29])  

# Crear la figura
plt.figure(figsize=(10, 6))

plt.style.use('seaborn-v0_8-pastel')

# Graficar las funciones
plt.plot(x, y_original, label='f(x) = 1 / (25x² + 1)', color='teal', linewidth=1.5)
plt.plot(x, y_lagrange_o3, label='Lagrange de orden 3', color='blue', linestyle='--',linewidth=1.5)
plt.plot(x, y_lagrange_o5, label='Lagrange de orden 5', color='orange', linestyle='--',linewidth=1.5)

# Etiquetas y detalles
plt.title('Interpolacion de lagrange')
plt.scatter(x_points, y_points, color='red', label='Puntos dados')
plt.xlabel('x', fontsize=13)
plt.ylabel('y', fontsize=13)
plt.axhline(0, color='gray', linewidth=0.8, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.8, linestyle='--')
plt.legend(fontsize=10)
plt.grid(True, linestyle=':', alpha=0.5)
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Estilo pastel de seaborn
plt.style.use('seaborn-v0_8-pastel')

# Rango de x
x = np.linspace(-10, 10, 1000)

# Función original y aproximaciones de Taylor
y_original = np.arctan(x)
y_taylor_0 = np.ones_like(x)
y_taylor_2 = (np.pi)/4 + 1/2*(x-1) - 1/4 * (x-1)**2
y_taylor_4 = y_taylor_2 + 1/12 * (x-1)**3
y_taylor_6 = y_taylor_4 - 1/40 * (x-1)**5 - 3/96 * (x-1)**6

# Crear figura
plt.figure(figsize=(12, 7))

# Graficar funciones con colores suaves y agradables
plt.plot(x, y_original, label='f(x) = arctan(x)', color='teal', linewidth=2)
plt.plot(x, y_taylor_0, label='Taylor orden 0', color='blue', linestyle='--', linewidth=2)
plt.plot(x, y_taylor_2, label='Taylor orden 2', color='skyblue', linestyle='-.', linewidth=2)
plt.plot(x, y_taylor_4, label='Taylor orden 4', color='salmon', linestyle=':', linewidth=2)
plt.plot(x, y_taylor_6, label='Taylor orden 6', color='orchid', linestyle='--', linewidth=2)

# Personalización
plt.ylim([-1.5, 1.5])
plt.title('Aproximaciones de la Serie de Taylor', fontsize=16)
plt.xlabel('x', fontsize=13)
plt.ylabel('y', fontsize=13)
plt.axhline(0, color='gray', linewidth=0.8, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.8, linestyle='--')
plt.legend(fontsize=10)
plt.grid(True, linestyle=':', alpha=0.5)
plt.tight_layout()


# Mostrar
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Rango de x
x = np.linspace(-1.1, 1.1, 1000)
y_original = np.arctan(x)
y_lagrange_o3 = (
   -(np.pi/4) * ((x*(x-1))/2) +
   (np.pi/4) * ((x*(x+1))/2)
)


x_points = np.array([-1,0,1])
y_points = np.array([-(np.pi/4), 0, (np.pi/4)])  

# Crear la figura
plt.figure(figsize=(10, 6))

plt.style.use('seaborn-v0_8-pastel')

# Graficar las funciones
plt.plot(x, y_original, label='f(x) = arctan(x)', color='teal', linewidth=1.5)
plt.plot(x, y_lagrange_o3, label='Lagrange de orden 2', color='blue', linestyle='--',linewidth=1.5)


# Etiquetas y detalles
plt.title('Interpolacion de lagrange')
plt.scatter(x_points, y_points, color='red', label='Puntos dados')
plt.xlabel('x', fontsize=13)
plt.ylabel('y', fontsize=13)
plt.axhline(0, color='gray', linewidth=0.8, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.8, linestyle='--')
plt.legend(fontsize=10)
plt.grid(True, linestyle=':', alpha=0.5)
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```
