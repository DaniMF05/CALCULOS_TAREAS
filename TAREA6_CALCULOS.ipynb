```python
import numpy as np
import matplotlib.pyplot as plt

# Estilo oscuro
plt.style.use('dark_background')

# Rango de x
x = np.linspace(-0.3, 0.3, 1000)

# Función original y aproximaciones de Taylor
y_original = 1 / (25 * x**2 + 1)
y_taylor_0 = np.ones_like(x)
y_taylor_2 = 1 - 25 * x**2
y_taylor_4 = 1 - 25 * x**2 + 312.5 * x**4
y_taylor_6 = 1 - 25 * x**2 + 312.5 * x**4 - ((4312500/720)*x**6)

# Crear figura
plt.figure(figsize=(12, 7))

# Graficar funciones con alto contraste
plt.plot(x, y_original, label='f(x) = 1 / (25x² + 1)', color='cyan', linewidth=2)
plt.plot(x, y_taylor_0, label='Taylor orden 0', color='white', linestyle='--', linewidth=2)
plt.plot(x, y_taylor_2, label='Taylor orden 2', color='lime', linestyle='-.', linewidth=2)
plt.plot(x, y_taylor_4, label='Taylor orden 4', color='orange', linestyle=':', linewidth=2)
plt.plot(x, y_taylor_6, label='Taylor orden 6', color='magenta', linestyle=':', linewidth=2)


# Personalización
plt.ylim([0, 1.1])
plt.title('Aproximaciones de la Serie de Taylor', fontsize=16, color='white')
plt.xlabel('x', fontsize=13, color='white')
plt.ylabel('y', fontsize=13, color='white')
plt.axhline(0, color='gray', linewidth=0.8, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.8, linestyle='--')
plt.legend(fontsize=10)
plt.grid(True, linestyle=':', alpha=0.5)
plt.tight_layout()

# Mostrar
plt.show()
```
