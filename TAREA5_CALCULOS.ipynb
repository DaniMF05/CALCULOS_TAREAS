```python
import math

# Definimos la función y su derivada
def f(x):
    return -x**3 - math.cos(x)

def f_prime(x):
    return -3 * x**2 + math.sin(x)

# Parámetros iniciales
x0 = -1
tol = 1e-6
max_iter = 100

print("Método de Newton-Raphson")
print(f"{'Iter. ,':<6}{'x_n ,':<15}{'f(x_n) ,':<15}{'f\'(x_n) ,':<15}{'x_{n+1} ,':<15}{'Error':<15}")

# Iteración manual
for i in range(1, max_iter + 1):
    fx = f(x0)
    fpx = f_prime(x0)
    
    if fpx == 0:
        print("Derivada cero. No se puede continuar.")
        break
    
    x1 = x0 - fx / fpx
    error = abs(x1 - x0)
    
    # Imprimir la fila de la tabla
    print(f"{i:<6},{x0:<15.8f},{fx:<15.8f},{fpx:<15.8f},{x1:<15.8f},{error:<15.8f}")
    
    if error < tol:
        print("\nConvergencia alcanzada.")
        break
    
    x0 = x1
```

```python
# EJERCICIO 5 - BISECCION

def f(x: float) -> float:
    return ( (math.tan( math.pi * x) - 6))



def biseccion(a: float, b: float, tol: float = 1e-5, max_iter: int = 11):
    if f(a) * f(b) >= 0:
        print("El método de bisección no es aplicable: f(a) y f(b) deben tener signos opuestos.")
        return None
    
    print(f(a))
    print(f(b))

    iteracion = 1
    while (b - a) / 2 > tol and iteracion < max_iter:
        c = (a + b) / 2
        fc = f(c)
        fb = f(b)
        minerror = abs(b-c)
        difer = fc * fb


        #print(f" {a:.6f}, {b:.6f}, {c:.6f}, {fc:.6f}, {minerror:.6f}, {difer:.6f}")
        #print(a,",", b,",",c,",",fc,",",minerror,",",difer)
        print(f"{iteracion:<6},{a:<15.8f},{b:<15.8f},{c:<15.8f},{fc:<15.8f},{minerror:<15.8f},{difer:<15.8f}")

        if abs(fc) < tol:
            return c
        elif f(a) * fc < 0:
            b = c
        else:
            a = c
        iteracion += 1

    return (a + b) / 2  # Valor aproximado de la raíz

biseccion(0, 0.48 )
```
