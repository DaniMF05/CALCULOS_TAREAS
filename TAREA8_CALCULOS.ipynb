```python
# minimos cuadrados

import numpy as np
from sympy import symbols, Eq, solve

x = [4, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3 ,6.8, 7.1 ]
y = [102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.5, 326.72]

suma_x = 0
suma_y = 0
suma_xy = 0
suma_x2 = 0
suma_x3 = 0
suma_x4 = 0
suma_x2_y = 0
suma_error_g1 = 0
suma_error_g2 = 0

for i in range(len(x)):
    suma_x = suma_x + x[i]
    suma_y = suma_y + y[i]
    suma_xy = suma_xy + x[i] * y[i]
    suma_x2 = suma_x2 + x[i]**2
    suma_x3 = suma_x3 + x[i]**3
    suma_x4 = suma_x4 + x[i]**4
    suma_x2_y = suma_x2_y + x[i]**2 * y[i]
    suma_error_g1 = (y[i] - (0.78 + 36.05 * x[i]))**2
    suma_error_g2 = (y[i] - (-0.0079 -0.000089 * x[i] + 6.46 * x[i]**2))**2

print("Suma de x:", suma_x)
print("Suma de y:", suma_y)
print("Suma de xy:", suma_xy)
print("Suma de x^2:", suma_x2)
print("Suma de x^3:", suma_x3)
print("Suma de x^4:", suma_x4)
print("Suma de x^2 * y:", suma_x2_y)
print()
print("Suma de errores grado 1:", suma_error_g1)
print()
print("Suma de errores grado 2:", suma_error_g2)


A = np.array([[10, 54.1],
              [541, 303.39]])

b = np.array([1958.39, 11361.76])

# Resolver el sistema para el grado 1
x_sol = np.linalg.solve(A, b)

print()
print("Soluci贸n grado 1:")
print(f"x = {x_sol[0]}")
print(f"y = {x_sol[1]}")

# Resolviendo el sistema para el grado 2

# Declarar variables
x, y, z = symbols('x y z')

# Definir ecuaciones
eq1 = Eq(10 * x + 11361.76* y + 303.39*z, 1958.39)
eq2 = Eq(541 * x + 303.39 * y + 1759.83*z, 11361.76)
eq3 = Eq(303.39 *x + 1759.83 * y +  10523.12* z, 67962.49)

# Resolver el sistema
soluciones = solve((eq1, eq2, eq3), (x, y, z), dict=True)

# Mostrar soluciones
print()
print("Soluciones grado 2:")
for i, sol in enumerate(soluciones, start=1):
    print(f"Soluci贸n {i}:")
    for var in sol:
        print(f"  {var} = {sol[var]}")
```
